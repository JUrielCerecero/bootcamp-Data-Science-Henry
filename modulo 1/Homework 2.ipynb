1) Abrir el archivo "Emisiones_CO2.csv" y cargar sus datos en un diccionario.

```python
import os
archivo = open('Emisiones_CO2.csv', 'r', encoding='latin1')
```

```python
dicc_emisiones = {  'cod_pais' : [],
                    'nom_pais' : [],
                    'region' : [],
                    'anio' : [],
                    'co2' : [],
                    'co2_percapita' : []}
```

```python
for a in archivo:
    a=a.strip()
    c=a.split('|') 
    dicc_emisiones['cod_pais'].append(c[0])
    dicc_emisiones['nom_pais'].append(c[1])
    dicc_emisiones['region'].append(c[2])
    dicc_emisiones['anio'].append(c[3])
    dicc_emisiones['co2'].append(c[4])
    dicc_emisiones['co2_percapita'].append(c[5])
for d in dicc_emisiones.keys():
    del dicc_emisiones[d][0]
for b in dicc_emisiones.keys():
   for d in range(0,len(dicc_emisiones[b])):
      if dicc_emisiones[b][d]=='':
         dicc_emisiones[b][d]='ND'
for b in ['co2','co2_percapita']:
    for c in range(0,len(dicc_emisiones[b])):
        if ',' in dicc_emisiones[b][c]:
            dicc_emisiones[b][c]=dicc_emisiones[b][c].replace('.','')
for b in ['co2','co2_percapita']:
    for c in range(0,len(dicc_emisiones[b])):
        if ',' in dicc_emisiones[b][c]:
            dicc_emisiones[b][c]=dicc_emisiones[b][c].replace(',','.')
for b in ['co2','co2_percapita']:
    for c in range(0,len(dicc_emisiones[b])):
        if not dicc_emisiones[b][c]=='ND':
            dicc_emisiones[b][c]=float(dicc_emisiones[b][c])
            
```
